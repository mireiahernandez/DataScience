## Pearson Correlation

To do a correlation plot, we must calculate first a **correlation matrix**. There is a pandas method *.corr()* for that. It computes the **linear or Pearson correlation**, which measures strength and direction of the linear relationship between two variables.

```python
corr = df[features_response].corr()
```

```python
corr.iloc[0:5,0:5]
```

Correlations range between -1 and 1, and each column has a correlation 1 with itself. So the matrix has 1 in diagonal and is symmetrical.

Now we may plot it:

```python
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, center=0)
```

- PAY_1 feature is the most strongly correlated feature to the response variable, so it's likely going to be the most predictive feature for our model.
- BILL_AMT features are highly correlated between each other. We will talk later about this.
- LIMIT_BAL is negatively correlated with the response variable.
- Apart form PAY_1 and LIMIT_BAL, the rest of the features have 0 correlation with the response variable.

Issues:
- Pearson correlation is only valid for **continuous data**, so it is technically not valid for our binary response variable problem. However, it still can give us a quick idea.
- Linear correlation is not an effective measure of the predictive power of all features. It only picks up linear relationships.

## Exercise 11: F-test and Univariate Feature Selection

**Univariate feature selection**: testing features one by one against reponse variable to see which ones have predictive power.

**ANOVA F-test**: f_classif from sklearn

The p-value answers the question: 'For the samples in the positive class, how likely is that the average value of this feature is the same as that of samples in the negative class?'

If p-value < 0.05, then probably this is a good feature because it will help us discriminate between positive and negative classes.

```python
from sklearn.feature_selection import f_classif
```

```python
X = df[features_response[:-1]].values
y = df[features_response[-1]].values
```

```python
X.shape
```

```python
y.shape
```

```python
F, pval = f_classif(X,y)
```

Build a df with each feature and their corresponding F statistic and p-value:

```python
f_test_df = pd.DataFrame({
    'Feature':np.array(features_response[:-1]),
    'F statistic': F,
    'p value': pval
})
```

```python
f_test_df
```

LIMIT_BAL and PAY_1 appear to be the most useful features. They have the smallest p-value, indicating the average values of these features between the positive and negative classes are significantly different.

If we want to retrieve only the 20% of features with lowest p-value, we can do this with **SelectPercentile** class. Or if we want to select the top k features **SelectKBest**.

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selector = SelectPercentile(f_classif, percentile=20)
```

```python
selector.fit(X,y)
```

To reduce X to the selected features:

```python
X_new = selector.transform(X)
```

```python
X_new.shape
```

If you want to get the names of the selected features:

```python
best_features_ix = selector.get_support()
```

```python
best_features_ix
```

```python
features = features_response[:-1]
```

```python
best_features = [features[counter] for counter in range(len(features)) if best_features_ix[counter]]
```

```python
best_features
```

However in this plot the only take-away is that the blue histogram is bigger than the red one, meaning that there are more accounts that don't default. We already knew this!

What we want to know is if different credit limit groups have different proportion of default/non-default accounts. Therefore we should normalize each histogram by its total area. We can do this with the keyword *density=True*.

Let's determine the number of bins that we want:

```python
df['LIMIT_BAL'].min()
```

```python
df['LIMIT_BAL'].max()
```

```python
bin_edges = list(range(0, 850000, 50000))
```

```python
bin_edges[-1]
```

```python
axes = plt.axes()
# alpha is transparency
axes.hist(df.loc[neg_mask, 'LIMIT_BAL'], bins=bin_edges, alpha=0.5, density=True, color='blue')
axes.hist(df.loc[pos_mask, 'LIMIT_BAL'], bins=bin_edges, alpha=0.5, density=True, color='red')
axes.set_xlabel('Credit limit')
axes.tick_params(axis='x', labelrotation=45)
axes.set_ylabel('Number of accounts')
axes.legend(['Not defaulted', 'defaulted'])
axes.set_title('Credit limits by response variable')
```

Now we can clearly see that the proportion of defaulted accounts is higher for accounts with lower credit limit.

However we have to multiply by 50000 (bin width) to get the proportion of accounts (because we have divided by area which is sum ( bin width * bin height ).

```python
axes = plt.axes()
# alpha is transparency
axes.hist(df.loc[neg_mask, 'LIMIT_BAL'], bins=bin_edges, alpha=0.5, density=True, color='blue')
axes.hist(df.loc[pos_mask, 'LIMIT_BAL'], bins=bin_edges, alpha=0.5, density=True, color='red')
axes.set_xlabel('Credit limit')
axes.tick_params(axis='x', labelrotation=45)
axes.set_ylabel('Proportion of accounts')
axes.legend(['Not defaulted', 'defaulted'])
axes.set_title('Credit limits by response variable')

# Fix the y labels
y_ticks = axes.get_yticks()
axes.set_yticklabels(np.round(y_ticks*50000, 2))
```

Accounts with credit limit lower than 150,000 are more likely to default, while is credit limit higher than 150,000 they are less likely to default.

## Exercise 14: Examining the Appropriateness of Features for Logistic Regression

Before we plotted the mean of the default rate as a function of the PAY variable. This is equivalent to plotting the "probability" of default rate.

However the relationship between probability of default rate and PAY is assumed to be log-linear, not linear, in logistic regression:

linear_model(X) = log( p/(1-p) )

Therefore we must plot the feature PAY versus the log(p/q):

```python
group_by_pay_mean_y
```

```python
p = group_by_pay_mean_y['default payment next month'].values
```

```python
q = 1-p
```

```python
p
```

```python
q
```

```python
odds_ratio = p/q
```

```python
log_odds = np.log(odds_ratio)
```

```python
feature_levels = group_by_pay_mean_y.index
```

```python
plt.plot(feature_levels, log_odds, '-x')
plt.xlabel('Values of PAY_1')
plt.ylabel('Log odds of default')
```

The relationship is not linear.

## From Logistic Regression Coefficients to Predictions Using the Sigmoid


```python
def sigmoid(x):
    Y = 1 / (1 + np.exp(-x))
    return Y
```

```python
X_sig = np.linspace(-7,7,141)
Y_sig = sigmoid(X_sig)
plt.plot(X_sig,Y_sig)
plt.yticks(np.linspace(0,1,11))
plt.grid()
plt.title('The sigmoid function')
plt.xlabel('$X = \Theta_0 + \Theta_1 X_1 + \Theta_2 X_2$')
plt.ylabel('$p$')
```

## Exercise 15: Linear Decision Boundary of Logistic Regression

Generate synthetic data (2 features X_1 and X_2 for 20 samples) for a binary classification problem.

```python
np.random.seed(seed=6)
```

```python
X_1_pos = np.random.uniform(low=1, high=7, size=(20,1))
```

```python
X_1_neg = np.random.uniform(low=3, high=10, size=(20,1))
```

```python
X_2_pos = np.random.uniform(low=1, high=7, size=(20,1))
```

```python
X_2_neg = np.random.uniform(low=3, high=10, size=(20,1))
```

Plot the data:

```python
plt.scatter(X_1_pos, X_2_pos, color='red', marker='x')
plt.scatter(X_1_neg, X_2_neg, color='blue', marker='x')
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
plt.legend(['Positive class', 'Negative class'])
```

Prepare data for Logistic Regression:

```python
X_1 = np.concatenate((X_1_pos, X_1_neg), axis=0)
```

```python
X_1.shape
```

```python
X_2 = np.concatenate((X_2_pos, X_2_neg), axis=0)
```

```python
X_2.shape
```

```python
X = np.concatenate((X_1, X_2), axis=1)
```

```python
X.shape
```

```python
X[0:3]
```

Create response variable (1st 20 samples are positive and last 20 samples are negatives):

```python
y = np.concatenate((np.ones((20,1)), np.zeros((20,1))), axis=0)
```

```python
y[0:5]
```

```python
y[-5:]
```

Build the linear regression model:

```python
from sklearn.linear_model import LogisticRegression
```

```python
my_lr = LogisticRegression(solver='liblinear')
```

```python
my_lr.fit(X,y)
```

```python
y_pred = my_lr.predict(X)
```

```python
positive_indices = [counter for counter in range(40) if y_pred[counter]==1]
```

```python
negative_indices = [counter for counter in range(40) if y_pred[counter]==0]
```

```python
positive_indices
```

We see that 17/20 positive samples were correctly predicted, and 18/20 negative samples were correctly predicted.

Now let's plot the predictions:

```python
plt.scatter(X_1_pos, X_2_pos, color='red', marker='x')
plt.scatter(X_1_neg, X_2_neg, color='blue', marker='x')
plt.scatter(X[positive_indices,0], X[positive_indices,1], s=150, marker='o',
            edgecolors='red', facecolors='none')
plt.scatter(X[negative_indices,0], X[negative_indices,1], s=150, marker='o',
            edgecolors='blue', facecolors='none')
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
plt.legend(['Positive class', 'Negative class', 'Positive predictions', 'Negative predictions'])
```

You see that the classifier struggles with data points close to where you can imagine that the decision boundary might be.

Let's plot the linear decision boundary. If we define X = theta0 + theta1X1 + theta2X2, then we know that p > 0.5 if and only if X > 0. Therefore the decision boundary (linear) is theta0 + theta1X1 + theta2X2 > 0, which is the same as:

X2 > - theta1/theta2 X1 - theta0/theta2

Let's get the coefficients:

```python
theta_1, theta_2 = my_lr.coef_[0]
```

```python
theta_0 = my_lr.intercept_[0]
```

```python
X.shape
```

```python
X_2_decbound = - theta_1/theta_2*X[:,0] - theta_0/theta_2
```

Repeat the previous plot but with the decision boundary:

```python
plt.scatter(X_1_pos, X_2_pos, color='red', marker='x')
plt.scatter(X_1_neg, X_2_neg, color='blue', marker='x')
plt.scatter(X[positive_indices,0], X[positive_indices,1], s=150, marker='o',
            edgecolors='red', facecolors='none')
plt.scatter(X[negative_indices,0], X[negative_indices,1], s=150, marker='o',
            edgecolors='blue', facecolors='none')
plt.plot(X[:,0], X_2_decbound)
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
plt.legend(['Positive class', 'Negative class', 'Positive predictions', 'Negative predictions'])
```

We can see that a linear decision boundary will never be able to classify perfectly. We could do feature engineering to allow for more complex, non-linear deicison boundaries. Or we could use a non-linear model such as random forest.
